##Rock, Paper, Scissors

####Game setup. 
Each game round consists two turns, the ﬁrst by the computer and the second by a human. The computer continues playing rounds until the human chooses to quit. 
• The computer chooses one of Rock, Paper, and Scissors, but keeps its choice secret. 
• The computer asks for the human’s input. 
• The human chooses one of Rock, Paper, and Scissors, or Quit.
• Unless the human quits, the computer ﬁgures out the result of the game, as follows: 
  – Rock smashes Scissors, so Rock beats Scissors.
  – Scissors can cut up paper, so Scissors beat Paper.
  – Paper covers Rock, so Paper beats Rock. 
  - If both players chose the same, it is a draw. The computer reports the result of this round. 
• If the human chooses to quit, the computer reports:
– the number of games played, and
– the number of times the human won.

####Computer’s brains. 
We will give the computer some smarts. It must be able to exploit any human biases. For example, if the human seems to prefer Rock, the computer should play Paper (which beats Rock). 


Our approach is to first make three different functions for the computer to decide
what it wants to "throw" in the game. One function will be for when the computer
has no valuable information about the user, and it will choose randomly amongst all three
options. The second option will be for when the computer knows the user has thrown
two options equally often, and it wants to randomly choose between the opposites
of those choices. The last function will be for when the computer knows a user has
thrown one choice the most. This function will return the opposite of that choice.

To make these functions work, we will need three helper functions. The first helper
function will find how many times the user has thrown his/her most often choice.
The second helper function will find how many of the choices have this maximum value.
This function will then help our third function, which will determine which function
the computer should call to decide its output.

Our penultimate function is called to determine who won the game - the computer
or the user. Our last function is used to play the game itself.

```python
import random

def all_random():
    """
    inputs: none
    outputs: a string - either rock, paper, or scissors, chosen randomly
    The computer chooses a choice (and keeps it secret)
    """
    throw = random.randint(1, 3)
    if throw == 1:
        computer_choice = "rock"
    elif throw == 2:
        computer_choice = "paper"
    elif throw == 3:
        computer_choice = "scissors"
        
    return computer_choice
```
